```python
import pandas as pd
import glob
import os
```

## Dataset Discovery

The dataset consists of multiple CSV files. The glob module is used to automatically identify all CSV files present in the dataset directory.

```python
files = glob.glob("../data/*.csv")

print("Total CSV files found:", len(files))
```

## Dataset Inspection

Before combining all CSV files, a sample dataset was inspected to understand the schema and verify consistency across files.

## Schema Validation

Before combining all datasets, the column structure of each CSV file was validated to ensure consistency.

```python
column_counts = []

for file in files:
    df = pd.read_csv(file)
    column_counts.append(len(df.columns))

print("Unique column counts:", set(column_counts))
```

```python
first_columns = list(pd.read_csv(files[0]).columns)

all_same = True

for file in files[1:]:
    current_columns = list(pd.read_csv(file).columns)

    if current_columns != first_columns:
        all_same = False
        print(f"Column mismatch found in {file}")
        break

print("All files have same columns:", all_same)
```

## Dataset Integration

After validating schema consistency across all CSV files, the datasets were combined into a single DataFrame using Pandas.

```python
all_dfs = []

for file in files:
    df = pd.read_csv(file)
    all_dfs.append(df)

combined_df = pd.concat(all_dfs, ignore_index=True)
```

## Handling Missing Values

The combined dataset was checked for missing values to identify incomplete records and determine appropriate cleaning steps.

```python
combined_df.isnull().sum()
```

```python
missing_values = combined_df.isnull().sum()

missing_values[missing_values > 0]
```

The dataset contains missing values in a few columns. These columns were identified before applying suitable data cleaning techniques.

```python
object_cols = combined_df.select_dtypes(include="object").columns

combined_df[object_cols] = combined_df[object_cols].fillna("Unknown")
```

```python
numeric_cols = combined_df.select_dtypes(include=["int64", "float64"]).columns

combined_df[numeric_cols] = combined_df[numeric_cols].fillna(0)
```

```python
combined_df.isnull().sum().sum()
```

Missing values in categorical columns were replaced with "Unknown", while missing values in numerical columns were replaced with 0. After cleaning, no missing values remained in the dataset.
