## Schema Validation

Before combining all datasets, the column structure of each CSV file was validated to ensure consistency.

```python
column_counts = []

for file in files:
    df = pd.read_csv(file)
    column_counts.append(len(df.columns))

print("Unique column counts:", set(column_counts))
```

```python
first_columns = list(pd.read_csv(files[0]).columns)

all_same = True

for file in files[1:]:
    current_columns = list(pd.read_csv(file).columns)

    if current_columns != first_columns:
        all_same = False
        print(f"Column mismatch found in {file}")
        break

print("All files have same columns:", all_same)
```
